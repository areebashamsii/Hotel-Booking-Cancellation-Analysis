# 4:Data Analysis:

```python
#Shows the percentage of canceled and non-canceled hotel reservations.
canceled_perc=df["is_canceled"].value_counts(normalize=True)
print(canceled_perc)
```

```python
# Plot the bar-graph of canceled and non-canceled hotel reservations.
plt.figure(figsize=(5,4))
plt.title("Hotel Reservations Status")
plt.bar(['Not_canceled','Canceled'],df["is_canceled"].value_counts(),edgecolor='k',width=0.7) 
plt.show()
           
```

```python
#Shows the graph of reservations status of both Hotels.
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel', hue='is_canceled',data=df,palette='Blues')
legend_labels, _=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different Hotels',size=20)
plt.xlabel('Hotels')
plt.ylabel('Number of Reservations')
plt.legend(['not_canceled','canceled'])
plt.show()
```

```python
#Shows the reservation status of the Resort Hotel in percentage.
resort_hotel=df[df["hotel"]== 'Resort Hotel']
resort_hotel["is_canceled"].value_counts(normalize=True)
```

```python
#Shows the reservation status of the City Hotel in percentage.
city_hotel=df[df["hotel"]== 'City Hotel']
city_hotel["is_canceled"].value_counts(normalize=True)
```

```python
#shows the adr(average daily rate ) on different dates of both the hotels.
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean
```

```python
#Shows the Bar grapgh of reservation Status for each month.
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month',hue='is_canceled' ,data=df)
plt.ylabel('Number of Reservations', fontsize=20)
plt.legend(['is_canceled','canceled'])
plt.show()
```

```python
#Shows the ADR status of Each month.
plt.figure(figsize=(16,8))
plt.title("ADR per Month",fontsize=20)
sns.barplot(x='month',y='adr',data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index(),palette="viridis")
plt.ylabel("ADR", fontsize=15)
plt.xlabel("MONTHS", fontsize=15)
plt.show()
```

```python
#Show the top 10 countries and their percentage of Canceled Reservations
canceled_data=df[df['is_canceled']==1]
top_10_countries=canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(16,8))
plt.title("Top 10 Countries with most Resevervation Cancellations" ,fontsize=15)
plt.pie(top_10_countries , autopct="%.2f", labels=top_10_countries.index)
plt.show()
```

```python
#show the count of each value of the column "market_segment".
df['market_segment'].value_counts()
```

```python
#Show the percentage of each value of the column "market_segment".
df['market_segment'].value_counts(normalize=True)
```

```python
canceled_data['market_segment'].value_counts(normalize=True)
```

```python
#Shows the canceled and not_canceled reservations data and its relation with adr using line graph.
#for canceled reservations:
canceled_df_adr=canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace=True)
canceled_df_adr.sort_values('reservation_status_date',inplace=True)

#for not-canceled reservations:

not_canceled_data=df[df['is_canceled'] == 0]
not_canceled_df_adr=not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df_adr.reset_index(inplace=True)
not_canceled_df_adr.sort_values('reservation_status_date',inplace=True)

#Plotting.
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'] , label='Canceled')
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'] , label='Not_canceled')
plt.legend(fontsize=15)
plt.show()

```

```python
#Extracting data from 2016 to 2017 because these years have consistent data.

not_canceled_df_adr=not_canceled_df_adr[(not_canceled_df_adr['reservation_status_date']>'2016')& (not_canceled_df_adr['reservation_status_date']<'2017-09')]
canceled_df_adr=canceled_df_adr[(canceled_df_adr['reservation_status_date']>'2016')& (canceled_df_adr['reservation_status_date']<'2017-09')]

#Plotting.
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'] , label='Canceled')
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'] , label='Not_canceled')
plt.legend(fontsize=15)
plt.show()
```
